1. Read the iris data(../data/iris.csv) and show first ten lines.(use readlines())

```python
import funcy
with open('../data/iris.csv', 'r') as f:
    raw_iris = f.readlines()
    
funcy.take(10, raw_iris)
```

2. From iris data, drop \n. And from this point, use the output as iris data.

```python
def drop_n(line):
    return line.replace('\n','')

iris = funcy.walk(drop_n, raw_iris)
```

3. From iris data, select the lines with condition that the last column is 'setosa' and show the first ten lines.

```python
def is_setosa(line):
    return line.split(',')[-1] == 'setosa'

funcy.take(10, funcy.filter(is_setosa, iris))
```

4. From iris data, sample 10 lines with replacement.

5. From iris data, make key-data with condition that the key is the last column and the value is the first column. And count the data per key.

```python
def species_first(line):
    return (line.split(',')[-1], line.split(',')[0])
    
funcy.walk_values(len, funcy.group_values(funcy.walk(species_first, iris)))
```

6. To the key-value data, sum-up based on the key(species).

```python
def convert_to_float(line):
    return list(funcy.walk(float, line))
    
funcy.walk_values(sum, funcy.walk_values(convert_to_float, funcy.group_values(funcy.walk(species_first, iris))))
```

7. To the key-value data, sort with descending by key and show the first 10 lines.

```python
funcy.take(10, sorted(funcy.walk(species_first, iris), key=lambda x: x[0], reverse=True))
```

8. By map() and reduce(), calculate the sum of all values of iris data except for species column.

```python
def concatenate_list(initial_list, line):
    final_list = list(initial_list)
    appender = final_list.append
    funcy.walk(appender, line)
    return final_list

# omit the last column
features = funcy.walk(lambda x: x.split(',')[:-1], iris)
# convert to float
float_features = funcy.walk(convert_to_float, features)
# concatenate the list
all_features_list = funcy.last(funcy.reductions(concatenate_list, float_features, []))
# sum up
sum(all_features_list)
```

9. Convert iris data to the form 
[{'sepal_length': val_1_1, 'sepal_width': val_2_1, 'petal_length': val_3_1, 'petal_width': val_4_1, 'species': val_5_1}, {'sepal_length': val_1_2, 'sepal_width': val_2_2, 'petal_length': val_3_2, 'petal_width': val_4_2, 'species': val_5_2},....]. (Apart from the species, everythong should be float type)

```python
def make_iris_dictionary(data_point):
    data_list = data_point.split(',')
    return {'sepal_length': float(data_list[0]), 'sepal_width': float(data_list[1]), 'petal_length': float(data_list[2]), 'petal_width': float(data_list[3]), 'species': data_list[4]}

iris_dictionary = funcy.walk(make_iris_dictionary, iris)
```

10. Calculate the ratio that the sepal_length > 5.0 to the whole iris data size.(Use iris_dictionary)

```python
len(list(funcy.filter(lambda x: x['sepal_length'] > 5.0, iris_dictionary))) / len(iris_dictionary)
```

11. Calculate the sepal_length values's average per species.

```python
funcy.group_by_keys(lambda x:x['species'], iris_dictionary)
```
